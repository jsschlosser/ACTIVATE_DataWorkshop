```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.ticker import MaxNLocator

# select research flight and load data and header files
df = pd.read_csv (r'ACTIVATE_20200229_data.csv')
df2 = pd.read_csv (r'ACTIVATE_20200229_column_labels.csv')
print (df2)

# select columns for x and y data
xi = 0
yi = 3

# get research flight date
a = df[df.columns[0]][:1]
txa = a[0].split(" ") 
txa = txa[0]

# x axis values 
x = df[df.columns[xi]]

if xi == 0:
    import numpy as np
    x2 = ["" for j in range(x.size)]
    for i in range(x.size):
        tx = x[i].split(" ") 
        x2[i] = tx[1]

# corresponding y axis values 
y = df[df.columns[yi]]
    
# create figure and subplot
fig,axes=plt.subplots(1,1)

# plotting the points  
axes.plot(x2, y)

# set number of x and y ticks
axes.xaxis.set_ticks(np.linspace(0,x.size,6))  
axes.yaxis.set_ticks(np.linspace(np.round_(min(y), decimals=-2, out=None),np.round_(max(y), decimals=-2, out=None),10))  

# naming the x axis 
a = df2[df2.columns[xi]]
plt.xlabel(a[xi]) 
# naming the y axis 
a = df2[df2.columns[yi]]
plt.ylabel(a[yi]) 
# gives a title to the Graph
plt.title(txa)
# function to show the plot 
plt.show() 
```
